La vitesse du son est la vitesse de phase dans la chaîne atomique donc 
$$c = \frac{\omega}{k}$$
Les trois branches acoustiques que nous étudierons seront les bandes 1, 2 et 3, qui sont les seules bandes acoustiques.
On calcule donc la pente sur chaque chemin choisi.

# Vitesse du son dans la direction Γ-X pour la bande 2

$$c = \frac{\Delta\omega}{\Delta k} = \frac{7.046919*10^{12}-0}{0.5391215*10^{10}-0} = 1307 m/s$$

# Vitesse du son dans la direction Γ-L pour la bande 3

$$c = \frac{\Delta\omega}{\Delta k} = \frac{(17.55082-1.220815)*10^{12}}{(5.173767-4.296821)*10^{10}} = 1862 m/s$$

# Vitesse du son dans la direction W-K pour la bande 1

$$c = \frac{\Delta\omega}{\Delta k} = \frac{(12.45613-11.79031)*10^{12}}{(2.64662-2.234906)*10^{10}} = 161.7 m/s$$

On remarque la vitesse du son est beaucoup plus importante aux alentours de Γ.
